```r
unemp_url <- "https://raw.githubusercontent.com/Arioniums/DSCI_100_125/main/Unemployment_Data_since_1987.csv"
unemp_path <- "Unemployment_Data_since_1987.csv"
download.file(unemp_url, destfile = unemp_path)
unemp_data <- read_csv(unemp_path)

#filtering for USA unemployment data, selecting for TIME and Value columns, and then filtering years to only 1997-2021
unemployment_data <- filter(unemp_data, LOCATION == "USA") |>
                    select(TIME, Value) |>
                    filter(between(TIME, 1987, 2021))

##renaming TIME column into year and Value column into annual_unemployment rate
names(unemployment_data)[names(unemployment_data) == 'TIME'] <- 'Year'
names(unemployment_data)[names(unemployment_data) == 'Value'] <- 'unemployment_rate_percentage'

unemployment_data
```

```r
#importing gdp growth rate data
gdp_url <- "https://raw.githubusercontent.com/Arioniums/DSCI_100_125/main/US_GDP_Growth.csv"
gdp_path <- "US_GDP_Growth.csv"
download.file(gdp_url, destfile = gdp_path)
g_data <- read_csv(gdp_path)

#reformatting the year values and filtering years to only 1997-2021
gdp_data <- select(g_data, Year, "GDP growth (annual %)") |>
                filter(between(Year, 1987, 2021))

#clarification of gdp growth rate unit (percentage)
names(gdp_data)[names(gdp_data) == 'GDP growth (annual %)'] <- 'gdp_growth_rate_percentage'

gdp_data
```
