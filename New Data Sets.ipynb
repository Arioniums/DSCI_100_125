```r
library(tidyverse)
```

```r
unemp_url <- "https://raw.githubusercontent.com/Arioniums/DSCI_100_125/main/Unemployment_Data_since_1987.csv"
unemp_path <- "Unemployment_Data_since_1987.csv"
download.file(unemp_url, destfile = unemp_path)
unemp_data <- read_csv(unemp_path)

#filtering for USA unemployment data, selecting for TIME and Value columns, and then filtering years to only 1997-2021
unemployment_data <- filter(unemp_data, LOCATION == "USA") |>
                    select(TIME, Value) |>
                    filter(between(TIME, 1987, 2021))

##renaming TIME column into year and Value column into annual_unemployment rate
names(unemployment_data)[names(unemployment_data) == 'TIME'] <- 'Year'
names(unemployment_data)[names(unemployment_data) == 'Value'] <- 'unemployment_rate_percentage'

unemployment_data
```

```r
#importing gdp growth rate data
gdp_url <- "https://raw.githubusercontent.com/Arioniums/DSCI_100_125/main/US_GDP_Growth.csv"
gdp_path <- "US_GDP_Growth.csv"
download.file(gdp_url, destfile = gdp_path)
g_data <- read_csv(gdp_path)

#reformatting the year values and filtering years to only 1997-2021
gdp_data <- select(g_data, Year, "GDP growth (annual %)") |>
                filter(between(Year, 1987, 2021))

#clarification of gdp growth rate unit (percentage)
names(gdp_data)[names(gdp_data) == 'GDP growth (annual %)'] <- 'gdp_growth_rate_percentage'

gdp_data
```

```r
#importing sixty/forty and forty/sixty portfolio data
portfolio_url <- "https://raw.githubusercontent.com/Arioniums/DSCI_100_125/main/Annual_Portfolio_Returns_and_Inflation_data.csv"                                          
portfolio_path <- "Annual_Portfolio_Returns_and_Inflation_data.csv"
download.file(portfolio_url, destfile = portfolio_path)

#selecting for year, 60/40_portfolio_return, 40/60_portfolio_return, and filtering years to only 1997-2021
portfolio_data <- read_csv(portfolio_path) |>
                                                select("Year", "Inflation", "Bogleheads Three Funds Portfolio", "60/40 Portfolio", "40/60 Portfolio", "Harry Browne Permanent Portfolio", "Gold") |>
                                                filter(between(Year, 1987, 2021))

#deleting the percentage unit in the column 60/40_portfolio_return and renaming the 60/40_portfolio_return into 60/40_portfolio_return_percentage
portfolio_data$'Inflation' = as.numeric(gsub("[\\%,]", "", portfolio_data$'Inflation'))
names(portfolio_data)[names(portfolio_data) == 'Inflation'] <- 'inflation_rate_percentage'

#deleting the percentage unit in the column 40/60_portfolio_return and renaming the 40/60_portfolio_return into 40/60_portfolio_return_percentage
portfolio_data$'Bogleheads Three Funds Portfolio' = as.numeric(gsub("[\\%,]", "", portfolio_data$'Bogleheads Three Funds Portfolio'))
names(portfolio_data)[names(portfolio_data) == 'Bogleheads Three Funds Portfolio'] <- 'Bogleheads Three Funds Portfolio_percentage'

portfolio_data$'60/40 Portfolio' = as.numeric(gsub("[\\%,]", "", portfolio_data$'60/40 Portfolio'))
names(portfolio_data)[names(portfolio_data) == '60/40 Portfolio'] <- '60/40 Portfolio_percentage'

portfolio_data$'40/60 Portfolio' = as.numeric(gsub("[\\%,]", "", portfolio_data$'40/60 Portfolio'))
names(portfolio_data)[names(portfolio_data) == '40/60 Portfolio'] <- '40/60 Portfolio_percentage'

portfolio_data$'Harry Browne Permanent Portfolio' = as.numeric(gsub("[\\%,]", "", portfolio_data$'Harry Browne Permanent Portfolio'))
names(portfolio_data)[names(portfolio_data) == 'Harry Browne Permanent Portfolio'] <- 'Harry Browne Permanent Portfolio_percentage'

portfolio_data$'Gold' = as.numeric(gsub("[\\%,]", "", portfolio_data$'Gold'))
names(portfolio_data)[names(portfolio_data) == 'Gold'] <- 'Gold_percentage'

#c stands for cleaned
portfolio_data
```

```r
#merging all cleaned datasets
data_merge_1 <- merge(
                    x = unemployment_data, 
                    y = gdp_data,
                    by = "Year")
data_merge_2 <- merge(
                    x = data_merge_1, 
                    y = portfolio_data,
                    by = "Year")

data_merge_2$max_portfolio_return_value <- do.call(pmax, data_merge_2[5:9])
data_merge_2$max_portfolio_return_name <- colnames(data_merge_2[5:9])[max.col(data_merge_2[5:9])]

data_merge_2
```

```r
new_final_data <- data_merge_2 |>
        select(unemployment_rate_percentage,
              gdp_growth_rate_percentage,
              inflation_rate_percentage,
              max_portfolio_return_value,
              max_portfolio_return_name)

new_final_data

final_dataframe <- collect(new_final_data)
write_csv(final_dataframe, "new_final_data.csv")
```
